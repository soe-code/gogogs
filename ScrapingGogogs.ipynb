```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
```

```python
### main ##############
### regular

df_reslt = pd.DataFrame(columns = ['maker-icon', 'shopID', 'shop-name', 'address', 'price', 'member-text', 'confirm-date'])

for i in range(1,48):

    page_num = 1
    #for page_num in range(1,3):
    while True:

        url = 'https://gogo.gs/ranking/' + str(i) + '?span=1&mode=0&submit=1&page=' + str(page_num)

        print(url)

        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html.parser')
        time.sleep(5)

        # maker-icon
        Icon_list = []
        for td in soup.find_all('td',class_='icon-td'):
            tes = td.find_all(class_='maker-icon')
            for a in tes:
                Icon_id = re.findall(r'.{5}maker_+.+', a['src'])
                Icon_list.append(Icon_id[0])
        
        # shop-ID
        tes = soup.find_all(attrs={"class": "shop-name"})
        GSshopID_list = []
        for a in tes:
            shopID = re.findall(r'\d+', a['href'])
            GSshopID_list.append(shopID[0])
        
        # GSname
        tes =soup.find_all(attrs={"class": "shop-name"})
        GSname_list = []
        for a in tes:
            GSname = a.text
            GSname_list.append(GSname)

        # GSadress
        tes =soup.find_all(attrs={"class": "address"})
        GSadress_list = []
        for a in tes:
            adress = a.text
            GSadress_list.append(adress)
        
        # GSprice
        tes =soup.find_all(attrs={"class": "price"})
        GSprice_list = []
        for a in tes:
            GSprice = a.text
            GSprice_list.append(GSprice)
        
        # GSmember
        tes = soup.find_all(attrs={"class": re.compile(".*-text")})
        GSmember_list = []
        for a in tes:
            member = a.text
            GSmember_list.append(member)
        
        # confirmdate
        tes =soup.find_all(attrs={"class": "confirm-date"})
        GSconfirmdate_list = []
        for a in tes:
            confirmdate = re.findall(r'2022/1..\d+', a.text)
            GSconfirmdate_list.append(confirmdate[0])


        # df_reslt
        df_kari = pd.DataFrame({ 'maker-icon': pd.Series(Icon_list[:20]), 'shopID': pd.Series(GSshopID_list),'shop-name':pd.Series(GSname_list)
                        ,'address':pd.Series(GSadress_list), 'price':pd.Series(GSprice_list), 'member-text':pd.Series(GSmember_list)
                        ,'confirm-date':pd.Series(GSconfirmdate_list)})

        df_reslt = pd.concat([df_reslt, df_kari])
        

        # next_page
        next_link = soup.find("i", class_='fa fa-angle-double-right') 
        if next_link is None:
            break
        else:
            page_num += 1


df_reslt
```

```python
df_reslt.reset_index(drop=True).to_csv('save/20221231_regular_gogogs.csv',index=True)
```
